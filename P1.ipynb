# D. Probabilities

#### The objective of this task is to assess your understanding of the conditional probability, but it could be also considered as a data exploration question.

Compute the probability that a randomly selected commune with elevation over 2000 is from the canton of Valais.

```python
df[df['Elevation'] > 2000]
```

```python
df[(df['Elevation']>=2000) & (df['Canton']=='VS')].shape[0]
```

```python
df[df['Elevation'] > 2000].Canton.value_counts(normalize=True)
```

Answer: 0.51

```python
import random 
```

```python
experiments = 10000
success = 0

df_reduced=df.loc[df['Elevation']>=2000]

for i in range(experiments):
    random_index = random.randrange(len(df_reduced))
    
    if df_reduced.iloc[random_index].Canton == 'VS' : success += 1
    
success/experiments
```
